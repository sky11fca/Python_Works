# Python Modules and Packages

## Modules

Besides the default modules, the Python Programming language comes by default. You can also import your own Python code to another. Achieveing M O D U L A R I T Y

The file `my_module.py` is an example of a function that returns the sum of 2 variables:

````python
import my_module
from Courses.WEEK9.MathOps.Simple import Arithmetic

print(my_module.Sum(10, 20))
````

Loading any module will cause to automatically execute the main code (The file's entry point).

In case the module is not in the same path as the Main program Will result an import error

Solution:
- Importing syntax: `import dir.dir.[...].module`
- Or: `sys.path += ['path/to/module']` (Require the sys module)

By default, each custom module has the following properties and methods

````python
print(dir(my_module))
````

````python
print(my_module.__file__) # the full path to the file that coresponds the module
print(my_module.__name__) # the name of the module
print(my_module.__package__) # name of the package
````

The `__doc__` attribute is used to write documentation for certain function, object

````python
print(my_module.__doc__)
````

Alternatively, the keyword `help` will output the entirety of information regarding that said module

````python
help(my_module)
````

The `__name__` attribute stores the name of the module. By default, executed in a main program returns `__main__`. But imported, will return the name of the module

It's a convenient approach to write `if __name__ == "__main__":` to avoid executing the code in the imported module.

## Packages

A grouping of Python Scripts is a package. Packages must be stored in a folder and each folder must contain an `__init__.py` file. This is the package's entry point

````python
# import MathOps.Simple.Arithmetic
from MathOps.Simple import Arithmetic as a

print(a.Add(10, 20))
````

````python
from MathOps.Simple import *

print(Arithmetic.Add(10, 20))
print(Bits.ShiftL(2, 3))
````

In order to use `from module import *` We must define the variable `__all__` inside the init file

If you want for the module/package to be available to all the scripts, copy paste the module or package forled to the Python Search path

- Windows: `<PythonPath>\Lib`
- Linux: `/usr/lib/<PythonVersion>`

The `importlib` library is used to import the module dynamically

-`import_module(module_name, package=None)` importing a module
-'reload(module)' reloading a module

````python
import importlib

m = importlib.import_module("my_module")
print(m.Sum(10, 20))
````

## Dynamic Code

The `exec` keyword is used for compiling/executing the python code dynamically

```
exec(code, [global], [local])
```

where global and local are lists of definitions executed depending on the scope

````python
exec("x=100")
print(x)
````

````python
exec("def num_sum(x,y): return x+y")
print(num_sum(10,20))
````

````python
s = "abcdefg"
exec("s2=s.upper()")
print(s2)
````
