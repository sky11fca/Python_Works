# Problem 1

```python
def reverse_vowels(s: str) -> str:
    VOWELS = "aeiouAEIOU"

    s_v = [c for c in s if c in VOWELS]
    replacements = {s_v[i]: s_v[-i-1] for i in range(len(s_v))}

    for i in range(len(s)):
        if s[i] in VOWELS:
            s = s[:i] + replacements[s[i]] + s[i+1:]

    return s
reverse_vowels("Hello Nica")
```

# Problem 2

```python
def analyze_scores(scores: list[tuple[str, int]]) -> dict:
    result = {"Average": 0.0, "Unique_scores": [], "top_scorer": None}

    average = 0
    unique_scores = set()
    max_score = scores[0][1]
    name_max_score = scores[0][0]

    for entries in scores:
        average += entries[1]
        unique_scores.add(entries[1])

        if entries[1] > max_score:
            max_score = entries[1]
            name_max_score = entries[0]

    result["Average"] = average / len(scores)
    result["Unique_scores"] = sorted(list(unique_scores))
    result["top_scorer"] = name_max_score

    return result

print(analyze_scores([("Alice", 90), ("Bob", 80), ("Michael", 95), ("Gabe", 80)]))
```

# Problem 3

```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, value):
        self.balance += value

    def withdraw(self, value):
        if value > self.balance:
            raise ValueError("Insufficient funds")
        self.balance -= value

    def __str__(self):
        return f"Balance: {self.balance}"
```

```python
acc = BankAccount(100)
acc.deposit(50)

try:
    acc.withdraw(200)
except ValueError as e:
    print(e)

print(acc)
```

# Problem 4

```python
fd = open("students.txt", "r")

entries = []

for line in fd:
    name, score = line.split(",")
    entries.append((name, int(score)))
fd.close()

max_score = entries[0][1]
max_scoring_student = entries[0][0]
average = 0.0
student_list = [student_name for student_name, score in entries]

for entry in entries:
    average += entry[1]
    if entry[1] > max_score:
        max_score = entry[1]
        max_scoring_student = entry[0]

open("results.txt", "w").write(
    f"Average score: {average / len(entries)}\n"
    f"Highest score: {max_score} by {max_scoring_student}\n"
    f"Students: {', '.join(student_list)}"
)
```
