# Exercise 1

Function that returns the first n numbers of the fibonacci sequence

```python
def fib(n, result=[], a=1, b=1):
    if len(result)==n:
        return result
    else:
        result.append(a)
        return fib(n, result, b, a+b)
```

```python
print(fib(5))
```

# Exercise 2

Function that given a list of numbers, returns a list of prime numbers

```python
def prime_list(list_nr):
    def is_prime(x):
        div = 0
        for i in range(1, x+1):
            if x%i==0:
                div+=1
        if div==2:
            return True
        else:
            return False

    result = [nr for nr in list_nr if is_prime(nr)]

    return result
```

```python
print(prime_list([2, 3, 4, 5, 6, 7]))
```

# Exercise 3

Given 2 lists: a, b compute the following: a union b, a intersected b, a-b, b-a

```python
def list_operation(a, b):
    union = list(set(a+b))
    intersection = list(set(a).intersection(b))
    dif_a_b = list(set(a).difference(b))
    dif_b_a = list(set(a).difference(b))

    return union, intersection, dif_a_b, dif_b_a
```

```python
print(list_operation([1, 2, 3], [3, 4, 5]))
```

# Exercise 4

Given a list of musical notes, a set of indexes and an offset. Write a function that returns a list that follows a song based off the indexes+offset

```python
def song(notes, indexes, offset):
    result = []
    for i in indexes:
        result.append(notes[(i+offset-1)%len(notes)])
    return result
```

```python
print(song(["do", "re", "mi", "fa", "sol"], [1, -3, 4, 2], 2))
```

# Exercise 5

Given a matrix, the function will return the matrix with the element 0 under the main diagonal

```python
def operation(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            if i>j:
                matrix[i][j]=0

    return matrix
```

```python
print(operation([[1, 2, 3], [4, 5, 6], [7, 8, 9]]))
```

# Exercise 2

Given a list of lists and a number, write a function that returns elements from the lists that appear n times

```python
def appearances(matrix, x):
    element_dict = dict()

    for x_list in matrix:
        for element in x_list:
            if element not in element_dict:
                element_dict[element] = 1
            else:
                element_dict[element] += 1

    result_list = [key for key, value in element_dict.items() if value == x]

    return result_list
```

```python
print(appearances([[1, 2, 3], [2, 3, 4], [4, 5, 6], [4, 1, "test"]], x=2))
```

# Exercise 7

Given a list of numbers, will return a tuple consisting of first number being the nr of palindrome numbers and other the maximum palindrome number

```python
def palindrome_lists(number):
    def is_palindrome(nr):
        return str(nr) == str(nr)[::-1]

    pal_list = [n for n in number if is_palindrome(n)]

    return (len(pal_list), max(pal_list))
```

```python
print(palindrome_lists([1234321, 1221, 12321, 232323, 55555]))
```

# Exercise 8

Given x (default value 1) a list of strings and a flag (true by default), write a function that does the following:

- If flag is true then will return a list of char lists consisting of character whose ascii value is divisible by x.
- Otherwise will return a list of lists whose character are not divisible by x

```python
def fun(word_list, x=1, flag=True):
    result = []
    if flag:
        for word in word_list:
            result.append([c for c in word if ord(c)%x==0])
        return tuple(result)
    else:
        for word in word_list:
            result.append([c for c in word if ord(c)%x!=0])
        return tuple(result)
```

```python
print(fun(["test", "hello", "lab002"], x=2, flag=False))
```

# Exercise 9

Given a matrix, representing the heights of the spectators in a stadium. The function will return a list of tuples consisting of coordinates for the players that cannot see the game

We say a viewer cannot see if `m[i][j] < m[i-1][j]`

```python
def set_viewer_cannot_see(matrix):
    result = []

    for i in range(1, len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j]<matrix[i-1][j]:
                result.append((i, j))

    return result
```

```python
print(set_viewer_cannot_see([[1, 2, 3, 2, 1, 1],
 [2, 4, 4, 3, 7, 2],
 [5, 5, 2, 5, 6, 4],
 [6, 6, 7, 6, 7, 5]]))
```

# Exercise 10

Function that returns a given matrix into a list of tuples that is it's transpose

```python
def transpose(matrix):
    result = []

    for i in range(len(matrix)):
        result.append(tuple(matrix[j][i] for j in range(len(matrix))))

    return result
```

```python
print(transpose([[1,2,3], [5,6,7], ["a", "b", "c"]]))
```

# Exercise 11

function that given a list of tuples, it will be sorted based off the 3rd character from the second element of the tuple

```python
def sorted_tuple(tuple_list):
    return sorted(tuple_list, key=lambda x: x[1][2])
```

```python
print(sorted_tuple([('abc', 'bcd'), ('abc', 'zza')]))
```

# Exercise 12

Given a list of words, the function will return a pair list of words that rhyme

```python
def group_by_rhyme(words):
    result = {}

    for word in words:
        ending = word[-2:]
        if ending not in result:
            result[ending] = [word]
        else:
            result[ending].append(word)

    return list(result.values())
```

```python
print(group_by_rhyme(['ana', 'banana', 'carte', 'arme', 'parte']))
```
