Given 2 integers from the console, compute the greatest common divisor

```python
from curses.ascii import isdigit

a = int(input())
b = int(input())

while b!=0:
    a, b = b, a%b

print(a)
```

# Exercise 2

Script that computes the number of vowels in a string

```python
VOWELS = "aeiouAEIOU"
sample = "Learning Python Momentum"

cnt=0

for c in sample:
    if c in VOWELS:
        cnt+=1

print(cnt)
```

# Exercise 3

Given 2 strings, compute the nr of appearence of string 2 in string 1

```python
s1 = "Linux is great, I love learning Linux"
s2 = "Linux"

s1_word_list = s1.split()

cnt=0

for word in s1_word_list:
    if word.lower() == s2.lower():
        cnt+=1

print(cnt)
```

# Exercise 4

Convert UpperCaseCamel to lowercase_with_underscores

```python
sample = "LearningPython"

result=[sample[0].lower()]

for i in range(1, len(sample)):
    if sample[i].isupper() :
        result.append("_")
        result.append(sample[i].lower())
    else:
        result.append(sample[i])

print("".join(result))
```

# Exercise 5

Function that validates if given number is palindrome

```python
def is_palindrome(nr):
    return str(nr) == str(nr)[::-1]
```

```python
# This should return true
print(is_palindrome(12321))
# This should be false
print(is_palindrome(12324))
```

# Exercise 6

Function that returns only digits

```python
def digit_only(s):
    result = [c for c in s if c.isdigit()]
    return "".join(result)
```

```python
print(digit_only("An apple is 123 USD")) # Damn Inflation
```

# Exercise 7

Function that returns how many 1 bits are in an integer number

```python
def bit_counter(nr):
    bin_nr = str(bin(nr))[2:]
    cnt=0
    for bit in bin_nr:
        if bit=='1':
            cnt+=1
    return cnt
```

```python
print(bit_counter(24))
```
